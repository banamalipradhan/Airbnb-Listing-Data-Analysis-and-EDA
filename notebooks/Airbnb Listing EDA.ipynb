**5. data analysis with EDA, Univariate Analysis**

```python
df["price"]    # price distributions 
```

```python
sns.boxplot(x="price",data=df)
plt.show()                      # identifying outliers
```

```python
data = df[df["price"]<1500]
sns.boxplot(x="price",data=data)
plt.show()                          # removing outliers
```

```python
plt.figure(figsize=(8,5))
sns.histplot(x="price",data=data,bins=70,kde=True)
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()                           # price distributions through histogram 
```

```python
plt.figure(figsize=(6,3))
sns.histplot(x="availability_365",data=df)
plt.xlabel("Availability")
plt.ylabel("Frequency")
plt.title("Availability Distribution")
plt.show()                           # Availability distributions through histogram 
```

```python
df.columns    # all feature names of the dataset
```

```python
data.groupby("neighbourhood_group")["price"].mean()    # average price for each neighbourhood group 
```

**Feature Engineering**

```python
data["price_per_bed"] = data["price"]/data["beds"]
data.head()                                          # creating new column as price_per_bed
```

```python
data.groupby("neighbourhood_group")["price_per_bed"].mean()    # average price per bed for each neighbourhood group 
```

**Bi-Variate Analysis** - To check how one variable depends on another variable

```python
data.columns    # all column names 
```

```python
sns.barplot(data=data,x="neighbourhood_group",y="price",hue="room_type")
plt.title("Neighbourhood by Price & Roomtype")
plt.show()                                                       # how price is depend on neighbourhood by using bar graph 
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=data,x="number_of_reviews",y="price",hue="neighbourhood_group")
plt.title("Locality & Review Dependency")
plt.show()                                           # relationship between number of reviews and price
```

```python
sns.pairplot(data=data,vars=["price","minimum_nights","number_of_reviews","availability_365"],hue="room_type")
plt.show()                                     # relationship between different variables by using pair plot 
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=data,x="longitude",y="latitude",hue="room_type")
plt.title("Geographical Distribution of Airbnb Listing")
plt.show()                                                    # geographical distribution of various room-types 
```
